```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df.shape
df.sample(7)
```

```python
df.info()
```

```python
df.describe()
```

```python
zero_columns = ["Pregnancies", "Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]

for i in zero_columns:
    zeros_count = (df[i] == 0).sum()
    print(f"{i}: {zeros_count}")
```

```python
for i in zero_columns:
    df[i] = df[i].replace(0,np.nan)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(zero_columns):
    sns.boxplot(y=df[col], ax=axes[i], color="lightblue")
    axes[i].set_title(col)

plt.suptitle("Box Plots — Outlier Detection", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
